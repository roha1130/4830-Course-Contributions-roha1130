```python

#The Table function is a fairly simple iterator that produces three yields for a given input function (we can call this the helper function).
#This code uses a function, in the text's case fnf(t)=cos⁡(t2) and calculates an accumulated value over a range of t values (0-4).
#Using a given step count, The total range is divided, and for each step, the function value for fnf(t) is multiplied by the step size delta(t) to calculate the change in y.
#Since y is the output change for each step, it is the value that we are accumulating, this can be useful to chart the change in something that can be represented by a growth function
#Ebola perhaps?
#The current values of delta(y), the accumulation, and t are printed for each iteration.
#Coming back to Activity 2, We can also plot this change, i.e. the accumulation, which will be implemented below.

#Function definition, this can be changed to any growth function, i.e. an Ebola outbreak
def fnf(t):
    return math.cos(t ** 2)

t_initial = 0
t_final = 4
number_of_steps = 2 ** 3
delta_t = (t_final - t_initial) / number_of_steps
t = t_initial
accumulation = 0

#Main iteration for given steps
for k in range(1, number_of_steps + 1):
    delta_y = fnf(t) * delta_t
    accumulation += delta_y
    t += delta_t
    print(f"Delta y: {delta_y:.6f}, Accumulation: {accumulation:.6f}, t: {t:.6f}")
```

## Activity 3

```python
import math
import matplotlib.pyplot as plt
```

```python
#The Plot function is similar to the Table function above.  It plots an accumulated delta(y) value and plots the accumulation over a certain number of iterations.
#The main difference is input parameters (a much higher step count), and the ouput as a plot as opposed to printing the actual values during each iteration.
#this is a visual representation of the change in delta(y).  The most important takeaway is that as we iterate through a growtih function fnf(), an accumulator records the values oy as they change, adding them to the total.
#This value is then plotted.

def fnf(t):
    return math.cos(t ** 2)

t_initial = 0
t_final = 4
number_of_steps = 400
delta_t = (t_final - t_initial) / number_of_steps
t = t_initial
accumulation = 0

t_values = [t_initial]
accumulation_values = [accumulation]

for k in range(1, number_of_steps + 1):
    delta_y = fnf(t) * delta_t
    accumulation += delta_y
    t += delta_t
    
    t_values.append(t)
    accumulation_values.append(accumulation)

plt.plot(t_values, accumulation_values)
plt.title('Accumulation of fnf(t) = cos(t^2)')
plt.xlabel('t')
plt.ylabel('Accumulation')
plt.grid(True)
plt.show()
```

```python
#Ebola Activity accumulation plots, a programmable version.
#Using the PLOT framework above.

def A(x):
    return 827 / ((10.5)**2 + (x - 55)**2)
def B(x):
    return 1400 / ((10)**2 + (x - 45)**2)
def C(x):
    return 1200 / ((10)**2 + (x - 45)**2)
def D(x):
    return 827 / ((10.5)**2 + (x - 44.4)**2)
def E(x):
    return 1000 / ((10.5)**2 + (x - 48)**2)
def F(x):
    return 827 / ((10.5)**2 + (x - 48)**2)

def plot_function(f, t_initial, t_final, number_of_steps, label, color):
    delta_t = (t_final - t_initial) / number_of_steps
    t = t_initial
    accumulation = 0

    t_values = [t_initial]
    accumulation_values = [accumulation]

    for k in range(1, number_of_steps + 1):
        delta_y = f(t) * delta_t
        plt.plot([t, t + delta_t], [accumulation, accumulation + delta_y], color=color, label=label if k == 1 else "")
        accumulation += delta_y
        t += delta_t

t_initial = 1
t_final = 100
number_of_steps = 400

plot_function(A, t_initial, t_final, number_of_steps, 'A(x)', 'blue')
plot_function(B, t_initial, t_final, number_of_steps, 'B(x)', 'green')
plot_function(C, t_initial, t_final, number_of_steps, 'C(x)', 'red')
plot_function(D, t_initial, t_final, number_of_steps, 'D(x)', 'purple')
plot_function(E, t_initial, t_final, number_of_steps, 'E(x)', 'black')
plot_function(F, t_initial, t_final, number_of_steps, 'F(x)', 'orange')

plt.title('Accumulation of Functions A, B, C, D, E, F')
plt.xlabel('x vals')
plt.ylabel('Accumulator')
plt.legend()
plt.grid(True)
plt.show()
```
