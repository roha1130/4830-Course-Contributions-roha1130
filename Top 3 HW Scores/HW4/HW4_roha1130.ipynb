```python
#Numerical Point Derivative

def Numerical_Point_Derivative(f, a, h, method='central'):     
    if method == 'central':
        slope = (f(a + h) - f(a - h)) / (2 * h)
    elif method == 'forward':
        slope = (f(a + h) - f(a)) / h
    else:
        raise ValueError("Invalid method. Choose 'central' or 'forward'.")

    print(f"Slope at a = {a} with h = {h} using {method} method: {slope}")
    return slope

#As an exmaple, we can use a, b, c, and d from #10 on pg 133.
def f(x):
    return 1/x
def g(x):
    return np.sin(7*x)
def i(x):
    return x**3
def s(x):
    return 2**x

#The central difference method remains the more effective method for approximating the derivative of. the input function, especially at smaller and smaller values of h.
#The central method is taking estimates both ahead of the inquiry point and behind, resulting in higher precision.  In terms of numerical point precision, it stabilizes to a much higher degree than the forward difference method.
#THe forward diff model is good enough for larger values of h, however at the scales that we need here, the central diff method wins out by a large margin, as we can see from the plot.
central_slope = Numerical_Point_Derivative(f, 2, 1e-5, method='central')
#central_slope = Numerical_Point_Derivative(g, 3, 1e-5, method='central')
#central_slope = Numerical_Point_Derivative(i, 200, 1e-5, method='central')
#central_slope = Numerical_Point_Derivative(s, 5, 1e-5, method='central')
central_slope = Numerical_Point_Derivative(f, 2, 1e-5, method='forward')
#central_slope = Numerical_Point_Derivative(g, 3, 1e-5, method='forward')
#central_slope = Numerical_Point_Derivative(i, 200, 1e-5, method='forward')
#central_slope = Numerical_Point_Derivative(s, 5, 1e-5, method='forward')
```

```python
#Numerical Function Derivative

def Numerical_Function_Derivative(f, x_1, x_2, points, h):     
    x_vals = []
    slope_update = []
    #Call back to the conitinuos slope plot function from earlier
    steps = (x_2 - x_1) / (points - 1)
    #We'll still be using the central difference method
    for i in range(points):
        x = x_1 + i * steps
        slope = (f(x + h) - f(x - h)) / (2 * h)
        x_vals.append(x)
        slope_update.append(slope)
        
    return x_vals, slope_update

#As an exmaple, we can use a, b, c, and d from #10 on pg 133.
def f(x):
    return 1/x
def g(x):
    return np.sin(7*x)
def i(x):
    return x**3
def s(x):
    return 2**x

x_vals, slopes = Numerical_Function_Derivative(f, 1, 3, 20, 1e-5)
plt.plot(x_vals, slopes)
plt.title('Derivative of f(x) = 1/x')
plt.xlabel('x')
plt.ylabel('Slope')
plt.grid(True)
plt.show()

x_vals, slopes = Numerical_Function_Derivative(g, 1, 3, 20, 1e-5)
plt.plot(x_vals, slopes)
plt.title('Derivative of g(x) = sin(7x)')
plt.xlabel('x')
plt.ylabel('Slope')
plt.grid(True)
plt.show()

# A little context for these two.  The plots appear the same from afar, but one is growing much quicker than the other (X^3), as you can see by the slope scale.
x_vals, slopes = Numerical_Function_Derivative(i, 1, 3, 20, 1e-5)
plt.plot(x_vals, slopes)
plt.title('Derivative of i(x) = x^3')
plt.xlabel('x')
plt.ylabel('Slope')
plt.grid(True)
plt.show()

x_vals, slopes = Numerical_Function_Derivative(s, 1, 3, 20, 1e-5)
plt.plot(x_vals, slopes)
plt.title('Derivative of s(x) = 2^x')
plt.xlabel('x')
plt.ylabel('Slope')
plt.grid(True)
plt.show()
```

## 5.

```python
def True_Deriv(f, x_1, x_2, points):
    steps = (x_2 - x_1) / (points - 1)
    x_vals = [x_1 + i * steps for i in range(points)]
    y_vals = [f(x) for x in x_vals]
    
    return x_vals, y_vals

def f(x):
    return np.log(x)
def g(x):
    return 7*np.cos(7*x)
def i(x):
    return 3*x**2
def s(x):
    return np.log(2)*2**x

x_vals, y_vals = True_Deriv(f, 1, 3, 20)
plt.plot(x_vals, y_vals)
plt.title('Derivative of f(x) = 1/x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

x_vals, y_vals = True_Deriv(g, 1, 3, 100)
plt.plot(x_vals, y_vals)
plt.title('Derivative of f(x) = 1/x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

x_vals, y_vals = True_Deriv(i, 1, 3, 20)
plt.plot(x_vals, y_vals)
plt.title('Derivative of f(x) = 1/x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

x_vals, y_vals = True_Deriv(s, 1, 3, 20)
plt.plot(x_vals, y_vals)
plt.title('Derivative of f(x) = 1/x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

#The results here can best be interpereted as a confirmation that the central diff method is very accurate when estimating the derivative of simple functions.
#The plots bellow, using the true derivates of the functions in #10, are nearly identical to the plots shown by the central diff estimate above.
#I used a higher point count for function b in order to plot the derivative smoothly.  This was really the lowest point count that I could use in order to smooth the line.
#The tricky part here is that the oscilations are too compressed on this scale.  A higher plot (point) count is required in order to get a smooth curve.  This is more resource intensive, but required.
#The others had smooth lines regardless at 20 points.
```
