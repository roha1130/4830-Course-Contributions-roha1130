```python
import matplotlib.pyplot as plt

tinitial = 0
tfinal = 40
S = 100000
I = 1
R = 0
D = 5
total_infected = 0
numberofsteps = 150
deltat = (tfinal - tinitial) / numberofsteps

times = []
S_values = []
I_values = []
R_values = []
a = 0.000015
t = tinitial

#The SIR Loop!
for k in range(numberofsteps):
    Sprime = -a * S * I
    Iprime = a * S * I - I / D
    Rprime = I / D
    
    deltaS = Sprime * deltat
    deltaI = Iprime * deltat
    deltaR = Rprime * deltat
    
    plt.plot([t, t + deltat], [S, S + deltaS], 'b-', lw=1)  # this is vs. time, as seen on the plot
    plt.plot([t, t + deltat], [I, I + deltaI], 'r-', lw=1)  # vs. time
    plt.plot([t, t + deltat], [R, R + deltaR], 'g-', lw=1)  # vs. time
    
    #increment by the change in time and SIR vars.
    t += deltat
    S += deltaS
    I += deltaI
    R += deltaR
    times.append(t)
    S_values.append(S)
    I_values.append(I)
    R_values.append(R)
    
    total_infected += I * deltat
    
print(f'Total infected during the outbreak: {total_infected:.0f}')

plt.xlabel('Time')
plt.ylabel('Pop')
plt.title('Measles outbreak model')
plt.legend(['S', 'I', 'R'])
plt.grid(True)
plt.show()
```

# Responses

The measles outbreak in this particular case would absolutely be considered an epidemic.  As we can see, the susceptible population goes down to 0 as we approach the end of the epidemic.  This means that the entire population had the infection at some point.

Nearly 2/3rds of the town will be infected by day 12, which means that there will be a peak in infections within two weeks.  The town will collectively gain immunity around the same time.  This means that every individual in the population will be exposed at one point or another.
